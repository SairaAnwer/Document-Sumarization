```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    document = fitz.open('/content/drive/MyDrive/Software_Engineering_10_Solutions_Manual (1).pdf')
    text = ""
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text()
    return text

# Example usage
pdf_text = extract_text_from_pdf('Software_Engineering_10_Solutions_Manual.pdf')
print(pdf_text[:2000])  # Print first 2000 characters for a quick check
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    return ' '.join(tokens)

# Example usage
preprocessed_text = preprocess_text(pdf_text)
print(preprocessed_text[:2000])  # Print first 2000 characters for a quick check
```

```python
from transformers import BartForConditionalGeneration, BartTokenizer

model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

def bart_summarize(text):
    inputs = tokenizer([text], max_length=1024, return_tensors='pt', truncation=True)
    summary_ids = model.generate(inputs['input_ids'], max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Example usage
summary = bart_summarize(preprocessed_text)
print(summary)
```
